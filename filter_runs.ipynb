```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
# Data files are formatted in a weird way, this function helps read them in
def read_sts_file(arr, cap=1000):
    df = pd.DataFrame()
    nm = 0
    for str in arr:
        if nm>cap:
            break
        with open(str) as f:
            data = json.load(f)
            df = pd.concat([df, pd.DataFrame(rec['event'] for rec in data)])
            nm+=1
    
    return df
```

```python
df = read_sts_file(["all_runs.json"])
df.head()
```

```python
df.shape[0]
```

```python
# filter
df = df[(df.is_endless == False) & (df.floor_reached <= 56) &
        (((df.victory == True) & (df.ascension_level >= 17)) |
         ((df.floor_reached >= 35) & (df.ascension_level >= 19)))]
df.shape[0]
```

```python
# train test split
df_train, df_test = train_test_split(df, train_size=0.8)
df_train.shape[0]
```

```python
df_test.shape[0]
```

```python
df_train.to_json("filtered_runs_train.json")
df_test.to_json("filtered_runs_test.json")
df.to_json("filtered_runs_full.json")
```
